```python
# Preprocessing function
def preprocess_text(text):
    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)
    # Convert to lowercase
    text = text.lower()
    return text

# Tokenization function
def tokenize_text(text):
    tokenizer = TreebankWordTokenizer()
    # Tokenize text into sentences
    sentences = sent_tokenize(text)
    # Tokenize each sentence into words
    words = [tokenizer.tokenize(sentence, convert_parentheses=True) for sentence in sentences]
    return words
```

```python
preprocessed_article = preprocess_text(sample["article"])
preprocessed_summary = preprocess_text(sample["highlights"])
# Tokenize article and summary
tokenized_article = tokenize_text(preprocessed_article)
tokenized_summary = tokenize_text(preprocessed_summary)
# Append tokenized article and summary to tokenized_data
(tokenized_article, tokenized_summary)
```
